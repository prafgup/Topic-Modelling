```python
import csv
import gensim
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from collections import defaultdict
import pprint
from gensim import corpora
import matplotlib.pyplot as plt
import os.path
```

```python
csvFile = open('state-of-the-union.csv')
csvReader = csv.reader(csvFile)
```

```python
stop_words = set(stopwords.words('english')) 
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
```

```python
def filterSpeech(speech):
    speech = speech.lower()
    new_speech =""
    for char in speech:
        if char not in punctuations:
            new_speech = new_speech + char
    speech = [w for w in new_speech.split() if not w in stop_words] 
    speech = " ".join(speech)
    return [x for x in speech.split() if not x.isdigit()]
```

```python
dataSet = {}
years=0
docs =0
for sp in csvReader:
    if sp[0] not in dataSet.keys():
        dataSet[sp[0]] = []
        years+=1
        dataSet[sp[0]].append(filterSpeech(sp[1]))
        docs+=1
print("years =" , years)
print("docs =" , docs)
```

```python
dataSet["1790"]
```

```python
frequency = defaultdict(int)
for year in dataSet.keys():
    for doc in dataSet[year]:
        for token in doc:
            frequency[token] += 1
```

```python
count = 0 
for elem in frequency.keys():
    if frequency[elem]>5:
        count+=1
print(count)
```

```python
for year in dataSet.keys():
    dataSet[year].append([token for token in dataSet[year][0] if frequency[token] > 5])
    dataSet[year].pop(0)

processed_corpus = [[token for token in dataSet[year][-1] if frequency[token] > 5]  for year in sorted(dataSet.keys())]
year_list = sorted(dataSet.keys())
```

```python
dictionary = corpora.Dictionary(processed_corpus)
corpus = [dictionary.doc2bow(text) for text in processed_corpus]
```

```python
from gensim import models
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
# corpus_tfidf = corpus
```

```python
processed_corpus[0]
```

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
def compute_coherence_values(dictionary, corpus_tfidf, processed_corpus, stop, start, step):
    coherence_values = []
    model_list = []
    for nt in range(start, stop, step):
        model = models.LsiModel(corpus_tfidf, num_topics=nt, id2word = dictionary)
        coherencemodel = CoherenceModel(model=model, texts=processed_corpus, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
        print("Coherence Value for num_topics=",nt," is ",coherencemodel.get_coherence())
    print(coherence_values)
    return coherence_values
```

```python
def plot_graph(processed_corpus,start, stop, step):
    coherence_values = compute_coherence_values(dictionary, corpus_tfidf,processed_corpus,                                                      stop, start, step)
    x = range(start, stop, step)
    plt.plot(x, coherence_values)
    plt.xlabel("Number of Topics")
    plt.ylabel("Coherence Score")
    plt.show()
```
